```python
with open("day_10.txt") as f:
    values = f.read().strip()
```

```python
values = values.replace('-', '═')
values = values.replace('|', '║')
values = values.replace('L', '╚')
values = values.replace('J', '╝')
values = values.replace('7', '╗')
values = values.replace('F', '╔')
```

```python
map = values.split('\n')

for y, line in enumerate(map):
    if "S" in line:
     x = line.index("S")
     break

start_pos = (x,y)
```

```python
def print_path(map, path=None):
    if path is None:
        for y, row in enumerate(map):
            for x, char in enumerate(row):
                    print(char, end="")
            print()
        return
        
    # Create a set for faster lookup
    path_set = set(path)

    for y, row in enumerate(map):
        for x, char in enumerate(row):
            if (x, y) in path_set:
                # Print the character in red
                print("\033[91m" + char + "\033[0m", end="")
            else:
                # Print the character in default color
                print(char, end="")
        print()  # Newline after each row
```

```python
def next_possible_moves(x, y):
    pipe = map[y][x]
    
    if pipe == "S":
        return None
    
    if pipe == "═":
        return [(x-1, y), (x+1, y)]
    elif pipe == "║":
        return [(x, y-1), (x, y+1)]
    elif pipe == "╚":
        return [(x, y-1), (x+1, y)]
    elif pipe == "╝":
        return [(x, y-1), (x-1, y)]
    elif pipe == "╗":
        return [(x-1, y), (x, y+1)]
    elif pipe == "╔":
        return [(x+1, y), (x, y+1)]

def get_pipe(x, y):
    return map[y][x]
```

```python
loop = []
loop.append(start_pos)
loop.append((start_pos[0], start_pos[1]+1)) # go right

while loop[-1] != start_pos:
    if len(loop) > 100000:
        break

    moves = next_possible_moves(*loop[-1])
    
    for move in moves:
        if move != loop[-2]:
            loop.append(move)
            #print(move, get_pipe(*move), loop[-1])
            break

print_path(map, loop)
```

```python
map2 = [list(m) for m in map]

for y, line in enumerate(map2):
    for x, col in enumerate(line):
        map2[y][x] = "#"
```

```python
def zoom_out_tile(tile):
    if tile == ".":
        return "   " + '\n' + "   " + '\n' + "   "
    
    if tile == "═":
        return "   " + '\n' + "###" + '\n' + "   "
    
    if tile == "║":
        return " # " + '\n' + " # " + '\n' + " # "
    
    if tile == "╚":
        return " # " + '\n' + " ##" + '\n' + "   "
    
    if tile == "╝":
        return " # " + '\n' + "## " + '\n' + "   "
    
    if tile == "╗":
        return "   " + '\n' + "## " + '\n' + " # "
    
    if tile == "╔":
        return "   " + '\n' + " ##" + '\n' + " # "

    if tile == "S":
        return "###" + '\n' + "#S#" + '\n' + "###"
    
```

```python
# zoom map out

zo_map = [["0" for _ in range(len(map2[0]) * 3)] for _ in range(len(map2) * 3)]

for y, line in enumerate(map):
    for x, tile in enumerate(line):
        zo_tile = zoom_out_tile(tile).split("\n")

        for ty, tline in enumerate(zo_tile):
            for tx, tchar in enumerate(tline):
                zo_map[y*3+ty][x*3+tx] = zo_tile[ty][tx]
```

```python
matrix = zo_map
width = len(zo_map)
height = len(zo_map[0])

def fill(x ,y, old, new):
    if x < 0 or x >= len(matrix[0]) or y < 0 or y >= len(matrix):
        return

    if matrix[y][x] != old:
        return

    matrix[y][x] = new

    fill(x+1, y, old, new)
    fill(x-1, y, old, new)
    fill(x, y+1, old, new)
    fill(x, y-1, old, new)


def fill_non_recursive(x, y, old, new):
    if x < 0 or x >= len(matrix[0]) or y < 0 or y >= len(matrix):
        return

    stack = [(x, y)]

    while stack:
        x, y = stack.pop()

        if matrix[y][x] != old:
            continue

        matrix[y][x] = new

        if x + 1 < len(matrix[0]):
            stack.append((x + 1, y))
        if x - 1 >= 0:
            stack.append((x - 1, y))
        if y + 1 < len(matrix):
            stack.append((x, y + 1))
        if y - 1 >= 0:
            stack.append((x, y - 1))


fill_non_recursive(0,0," ", "O")
```
